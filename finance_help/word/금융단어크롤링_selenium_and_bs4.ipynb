```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import os
```

```python
source_code = requests.get('https://fine.fss.or.kr/main/fin_tip/dic/financedic.jsp').text
```

```python
source_code
```

```python
html = BeautifulSoup(source_code,"html.parser")
word_list = html.select('.dic_result_list')
word = html.select('.dic_result_list dt')
meaning = html.select('.dic_result_list dd')

word_result = []
meaning_result=[]

for w in word:
  w.get_text()
  w = re.compile('[^ㄱ-ㅣ가-힣]+').sub("",str(w))
  word_result.append(w)
for m in meaning:
  m = str(m.get_text())
  m = m.split('.')[0]
  meaning_result.append(m)

df = pd.DataFrame(word_result,columns=["word"])
df['meaning'] = pd.DataFrame(meaning_result)
```
